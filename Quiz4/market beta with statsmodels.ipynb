```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_excel('beta.xlsx')
data
```

```python
data['일자'] = pd.to_datetime(data['일자'])
data.set_index('일자', inplace=True)
data = data.dropna()
```

```python
#선형 회귀
X = sm.add_constant(data['시장지수 초과수익률'])  
y = data['종목 초과수익률']  

#선형 회귀 모델 : sm.OLS
#fit() : 회귀모델 적합 시키는 것 
model = sm.OLS(y, X).fit()
beta = model.params['시장지수 초과수익률']
print(f"Beta: {beta}")
```
